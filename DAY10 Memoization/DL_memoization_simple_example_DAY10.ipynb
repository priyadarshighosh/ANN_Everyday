# Without Using MEMOIZATION

Fibonaci Series Without Using Memoization

```python
#FIBONAACI SERIES

import time
def fib(n):
    if n==0:
        return 0
    elif n==1:
        return 1
    else:
        return fib(n-1)+fib(n-2)
```

Calculating time  it takes to calculate the fibonaci series

```python
#TIME CALCULATION with the help of time seires

start = time.time()
print(fib(37))
end = time.time()
print(end-start)
```

# USING MEMOIZATION

We will see using memoization how fast the calculations are done .

```python
#fib
import time
def fib(n,d):
  if n in d:
    return d[n]        #if  we already have the number in the dictionary , we going to return that
  else:
    d[n]=fib(n-1,d)+fib(n-2,d)       # else we calculate return and store it in the dictionary
    return d[n]
```

Calculating the time it takes to calculating  fib series till the n th number

```python
#TIME CALCULATION with the help of time seires

start = time.time()

d={0:0,1:1}

print(fib(100,d))
end = time.time()
print(end-start)
```
