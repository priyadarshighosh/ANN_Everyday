# When alpha is 0.9 - We focus more on the current value
# BETA = ( 1 - ALPHA ) = 0.1

```python
x1=df['meantemp'].ewm(alpha=0.9).mean()       #beta value is 0.1  that means 0.1 weight to previous value
x1
```

```python
df['ewma']=x1                       # exponentially  weighted moving average
df.head()
```

```python
plt.scatter(df['date'],df['meantemp'],color='red')
plt.plot(df['date'],x1,color='blue', label='alpha=0.9')

plt.title('Mean Temperature')
plt.xlabel('Date')
plt.ylabel('Mean Temperature')
plt.legend()
plt.show()
```
