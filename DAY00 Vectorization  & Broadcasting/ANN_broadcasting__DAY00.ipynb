```python
import numpy as np
                                              # Printing the 2D array broadcasting example
A = np.array([[56.0 , 0.0 , 4.4 , 68.0 ],
             [1.2 , 104.0 , 52.0 , 8.0],
             [1.8 , 135.0 , 99.0 , 0.9]])

print(A)
```

```python
cal  = A.sum(axis=0)                        # We are finding the sum of the columns , thats why we did axis=0
print(cal)                                  # calculating calories in each column
```

```python
percentage = 100*A/cal.reshape(1,4)             # percentage of  carbs  proteins and fat are the rows
                                            # in each of the food - Apples , beef , eggs and Potatoes in the column
print(percentage)
```

We had a (3,4) matrix and we divided it by (1,4) matrix
