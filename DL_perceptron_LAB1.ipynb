```python
x =df.iloc[:,0:2]
y = df.iloc[:,-1]
```

```python
from sklearn.linear_model import Perceptron     # we are importing perceptron from the linear model
                                                 # linaer model is like a class of scikit lean
p = Perceptron()
```

```python
p.fit(x,y)   # we are fitting x y to the p function
```

```python
p.coef_  #to find W1 and W2
```

```python
p.intercept_   #to find the B
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(x.values, y.values, clf=p , legend=2)  #obviously this is not accurate it is just to show how a perceptron works and not how it should
                                                             #more over the amount of data we have is pretty so ya
```
